```python
import sys
sys.path.append('../src')

import numpy as np
import cvxpy as cp
from dbcp import BiconvexProblem

np.random.seed(10015)
```

```python
A = np.random.randn(10, 20)
X = cp.Variable((10, 5), nonneg=True)
Y = cp.Variable((5, 20), nonneg=True)

obj = cp.Maximize(-cp.sum_squares(X @ Y - A))
constr = [X >= 0, Y >= 0]
X.project_and_assign(np.random.standard_normal(X.shape))
Y.project_and_assign(np.random.standard_normal(Y.shape))
prob = BiconvexProblem(obj, ([X], [Y]), constr)
prob.is_dbcp()
```

```python
prob.solve()
```

```python
prob.status
```

```python
prob._x_prob.status
```

```python
x = cp.Variable((3, 3), PSD=True)
x.project_and_assign(np.random.randn(3, 3))
y = cp.Variable((3))

A = np.random.normal(size=(3, 3))
b = np.random.normal(3)

obj = cp.Minimize(cp.sum_squares(A @ x @y - b))
constr = [x >= 0]
prob = BiconvexProblem(obj, [[x], [y]], constr)
prob.is_dbcp()
```
